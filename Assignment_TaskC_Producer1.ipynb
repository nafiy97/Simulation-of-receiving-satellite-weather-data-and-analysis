```python
from time import sleep
from json import dumps
from kafka import KafkaProducer
import random
import datetime as dt

 
def publish_message(producer_instance, topic_name, data):
    try:
        producer_instance.send(topic_name, value=data)
        print('Message published successfully. ')
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                                  value_serializer=lambda x:dumps(x).encode('ascii'),
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
if __name__ == '__main__':
   
    topic = 'AssignmentB'
    print('Publishing records..')
    producer01 = connect_kafka_producer()

a=0   
import csv
import os
climate_reader = csv.reader(open('climate_streaming.csv','r'))

climate_list=[]
for line in climate_reader:
    climate_list.append(line)
while a==0: 
    i=random.randrange(1,len(climate_list))
    data={}
    time = str(dt.datetime.now().strftime("%Y-%m-%dT%H:%M:%S.000Z"))
    data["send_id"]="producer1"
    data["create_time"]=time
    data["latitude"]=float(climate_list[i][0])
    data["longitude"]=float(climate_list[i][1])
    data["air_temperature_celcius"]=int(climate_list[i][2])
    data["relative_humidity"]=climate_list[i][3]
    data["windspeed_knots"]=climate_list[i][4]
    data["max_wind_speed"]=climate_list[i][5]
    data["precipitation"]=climate_list[i][6]
    publish_message(producer01, topic, data) 
    sleep(5)
  
```
