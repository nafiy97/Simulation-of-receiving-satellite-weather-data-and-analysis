```python
from time import sleep
from json import dumps
from kafka import KafkaProducer
import random
import datetime as dt

 
def publish_message(producer_instance, topic_name, data):
    try:
        producer_instance.send(topic_name, value=data)
        print('Message published successfully. ')
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                                  value_serializer=lambda x:dumps(x).encode('ascii'),
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
if __name__ == '__main__':
   
    topic = 'AssignmentB'
    print('Publishing records..')
    producer03 = connect_kafka_producer()

    
import csv
import os
TERRA_reader = csv.reader(open('hotspot_TERRA_streaming.csv','r'))
a=0
TERRA_list=[]
for line in TERRA_reader:
    TERRA_list.append(line)
while a==0:
    i=random.randrange(1,len(TERRA_list))
    data={}
    time = str(dt.datetime.now().strftime("%Y-%m-%dT%H:%M:%S.000Z"))
    data['send_id']="producer3"
    data["create_time"]=time
    data["latitude"]=float(TERRA_list[i][0])
    data["longitude"]=float(TERRA_list[i][1])
    data["confidence"]=int(TERRA_list[i][2])
    data["surface_temperature_celcius"]=int(TERRA_list[i][3])
    publish_message(producer03, topic, data) 
    sleep(random.randrange(10,30))
```
