```python
import pymongo
from pymongo import MongoClient

client = MongoClient ()

# get database
db=client.fit5148_assignment_db 
# get collection
climate_collection = db.climate_collection
hotspot_collection = db.hotspot_collection
```

```python
import csv
import os
hotspot_reader = csv.reader(open('hotspot_historic.csv','r'))
climate_reader = csv.reader(open('climate_historic.csv','r'))

hotspot_list=[]
climate_list=[]
for line in hotspot_reader:
    hotspot_list.append(line)
for line in climate_reader:
    climate_list.append(line)    

hotspot_historic=[]
climate_historic=[]

#retrieve the hotspot_list as a list of dictionary
for i in range(1,len(hotspot_list)):
    demo_dic=dict()
    demo_dic['latitude']=hotspot_list[i][0]
    demo_dic['longitude']=hotspot_list[i][1]
    demo_dic['datetime']=hotspot_list[i][2]
    demo_dic['confidence']=int(hotspot_list[i][3])
    demo_dic['date']=hotspot_list[i][4]
    demo_dic['surface_temperature_celcius']=int (hotspot_list[i][5])

    #demo_dic['_id']=i
    hotspot_historic.append(demo_dic)
    

#retrieve the climate_list as a list of dictionary
for i in range(1,len(climate_list)):
    demo_dic=dict()
    demo_dic['station']=climate_list[i][0]
    demo_dic['date']=climate_list[i][1]
    demo_dic['air_temperature_celcius']=int(climate_list[i][2])
    demo_dic['relative_humidity']=climate_list[i][3]
    demo_dic['windspeed_knots']=climate_list[i][4]
    demo_dic['max_wind_speed']=climate_list[i][5]
    demo_dic['precipitation']=climate_list[i][6]
    climate_historic.append(demo_dic)
```

```python
from pprint import pprint

climate_collection = db.climate_collection
hotspot_collection = db.hotspot_collection

for lines in hotspot_historic:
    hotspot_collection.insert_one(lines)
for lines in climate_historic:
    pprint(lines)
    climate_collection.insert_one(lines)
```

```python
from pprint import pprint
result=climate_collection.find({"date":"10/12/2017"})
                              
for i in result:
    pprint(i)
```

```python
result=hotspot_collection.find({"$and":[{"surface_temperature_celcius":{"$gte":65}},
                                     {"surface_temperature_celcius":{"$lte":100}}]},
                                {"surface_temperature_celcius":1,"_id":0,"latitude":1,'longitude':1,'confidence':1}
                                
                              )
for i in result:
    pprint(i)
```

```python
result=climate_collection.aggregate([{
                                       "$lookup":
                                             {
                                                "from": "hotspot_collection",
                                                "localField":"date",
                                                "foreignField" :"date",
                                                "as": "surface_temperature"
                                               }},
                                      {"$match":{"date":{"$in":["15/12/2017","16/12/2017"]}}},
                                      {"$project":{"_id":0,"date":1,"surface_temperature.surface_temperature_celcius":1,
                                       'air_temperature_celcius':1,'relative_humidity':1,'max_wind_speed':1 }}
                                    ])
for i in result:
    pprint(i)
```

```python
result=hotspot_collection.aggregate([{
                                       "$lookup":
                                             {
                                                "from": "climate_collection",
                                                "localField":"date",
                                                "foreignField" :"date",
                                                "as": "air_temperature"
                                               }},
                                      {"$match":{"confidence":{"$gte":80,"$lte":100}}},
                                      {"$project":{"_id":0,"datetime":1,"air_temperature.air_temperature_celcius":1,
                                       'surface_temperature_celcius':1,'confidence':1 }}
                                    ])
for i in result:
    pprint(i)
```
